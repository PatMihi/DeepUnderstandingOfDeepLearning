```python
# import libraries
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

# Create the data and the model

```python
N = 100
x = torch.randn(N,1)
y = x + torch.randn(N,1)

# and plot
plt.plot(x,y,'s')
plt.show()
```

```python
train_size = int(0.8 * N)
```

```python
# build model
ANNreg = nn.Sequential(
    nn.Linear(1,1),  # input layer
    nn.ReLU(),       # activation function
    nn.Linear(1,1)   # output layer
    )
```

```python
# model meta-parameters

learningRate = .05

# loss function
lossfun = nn.MSELoss()

# optimizer
optimizer = torch.optim.SGD(ANNreg.parameters(),lr=learningRate)
```

# Select data for training

```python
# select training data (note the hard-coded N!)
trainidx  = np.random.choice(range(N),train_size,replace=False) # random indices
trainBool = np.zeros(N,dtype=bool) # initialize vector of Falses'
trainBool[trainidx] = True # set selected samples to True

# show the sizes
print(x[trainBool].shape)
print(x[~trainBool].shape)
```

# Train the model

```python
numepochs = 500
losses_train = np.zeros(numepochs)
losses_test  = np.zeros(numepochs)
for epochi in range(numepochs):

  # forward pass
  yHat = ANNreg(x[trainBool])

  # compute loss
  loss = lossfun(yHat,y[trainBool])
  losses_train[epochi] = loss.item()
  losses_test[epochi] = lossfun(ANNreg(x[~trainBool]),y[~trainBool]).item()
  

  # backprop
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
# report the losses

# compute losses of the TEST set
predYtest = ANNreg(x[~trainBool])
testloss = (predYtest-y[~trainBool]).pow(2).mean()

# print out final TRAIN loss and TEST loss
print(f'Final TRAIN loss: {loss.detach():.2f}')
print(f'Final TEST loss: {testloss.detach():.2f}')
```

```python
### plot the data

# predictions for final training run
predYtrain = ANNreg(x[trainBool]).detach().numpy()

# now plot
plt.plot(x,y,'k^',label='All data')
plt.plot(x[trainBool], predYtrain,
         'bs',markerfacecolor='w',label='Training pred.')
plt.plot(x[~trainBool],predYtest.detach(),
         'ro',markerfacecolor='w',label='Test pred.')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

# Train the model

```python
numepochs = 500
for epochi in range(numepochs):

  # forward pass
  yHat = ANNreg(X_train)

  # compute loss
  loss = lossfun(yHat,y_train)

  # backprop
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
# report the losses

# compute losses of the TEST set
predYtest = ANNreg(X_test)
testloss = (predYtest-y_test).pow(2).mean()

# print out final TRAIN loss and TEST loss
print(f'Final TRAIN loss: {loss.detach():.2f}')
print(f'Final TEST loss: {testloss.detach():.2f}')
```

```python
### plot the data

# predictions for final training run
predYtrain = ANNreg(X_train).detach().numpy()

# now plot
plt.plot(x,y,'k^',label='All data')
plt.plot(X_train, predYtrain,
         'bs',markerfacecolor='w',label='Training pred.')
plt.plot(X_test,predYtest.detach(),
         'ro',markerfacecolor='w',label='Test pred.')
plt.legend()
plt.show()
```
